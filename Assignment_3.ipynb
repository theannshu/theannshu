# **4.	Check and remove NULL values using dropna() and fillna() (with mean, median, and mode).**

# **of Dataset 1**

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/dataset 1/ufc-fighters-statistics.csv")
null_values = df.isnull().sum()
print("Null values in each column :\n", null_values)
df_dropped = df.dropna()

df_filled_mean = df.fillna(df.mean())
df_filled_median = df.fillna(df.median())
df_filled_mode = df.fillna(df.mode().iloc[0])

print("\nDataFrame after dropping NULL values:\n", df_dropped)
print("\nDataframe after filling NULL values with mean:\n", df_filled_mean)
print("\nDataframe after fillimg NULL values with median:\n", df_filled_median)
print("\nDataframe after filling NULL values with mode:\n", df_filled_mode)
```

# **Of dataset 2**

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/dataset 2/player_data_03_22.csv")
null_values = df.isnull().sum()
print("Null values in each column :\n", null_values)
df_dropped = df.dropna()

df_filled_mean = df.fillna(df.mean())
df_filled_median = df.fillna(df.median())
df_filled_mode = df.fillna(df.mode().iloc[0])

print("\nDataFrame after dropping NULL values:\n", df_dropped)
print("\nDataframe after filling NULL values with mean:\n", df_filled_mean)
print("\nDataframe after fillimg NULL values with median:\n", df_filled_median)
print("\nDataframe after filling NULL values with mode:\n", df_filled_mode)
```
